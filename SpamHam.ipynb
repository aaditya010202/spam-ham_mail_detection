```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('SMSSpamCollection',sep='\t',names=['label','msg'])
```

```python
df
```

```python
df.info()
```

```python
lower=[]
for i in df.msg:
  lower.append(" ".join([word.lower() for word in i.split()]))
lower
```

```python
alpha=[]
for i in lower:
  alpha.append(" ".join(word for word in i.split() if word.isalpha()))
alpha
```

```python
tokenize=[]
for i in alpha:
  tokenize.append([word for word in i.split()])
tokenize
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import *
stopword=nltk.corpus.stopwords.words('english')
without_stopword=[]
for i in tokenize:
  without_stopword.append([word for word in i if word not in stopword])
without_stopword
```

```python
final=[]
from nltk.stem import PorterStemmer
ps=PorterStemmer()
for i in without_stopword:
  final.append(" ".join([ps.stem(word) for word in i]))
final
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
bow_transformer=CountVectorizer().fit(final)
print(len(bow_transformer.vocabulary_))
```

```python
bow_transformer.vocabulary_
```

```python
bow=bow_transformer.transform(final)
print(bow)
print(bow.shape)
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
tfidf_transformer=TfidfTransformer().fit(bow)
tfidf=tfidf_transformer.transform(bow)
print(tfidf)
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB().fit(tfidf,df['label'])
```

```python
model.predict(tfidf[2])
```

```python
from sklearn.model_selection import train_test_split
msg_train, msg_test, label_train, label_test = train_test_split(df['msg'],df['label'],test_size=0.2)
```

```python
msg_test
```

```python
from sklearn.pipeline import Pipeline
pipeline = Pipeline([
    ('bow',CountVectorizer()),
    ('tfidf',TfidfTransformer()),
    ('model',MultinomialNB())
])
```

```python
pipeline.fit(msg_train,label_train)
```

```python
predictions=pipeline.predict(msg_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(predictions,label_test)
```
